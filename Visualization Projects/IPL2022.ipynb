# **TOSS DECISION TRENDS**

```python
sns.countplot(x = df["toss_decision"],palette = "rainbow")
plt.title("Toss Decision Trends")
```

# **TOSS WINNER VS MATCH WINNER**

```python
count = df[df["toss_winner"] == df["match_winner"]]["match_winner"].count()
print(f"out of 74 matches there were only {count} matches where the toss winner and the match winner was same team")
```

```python
percentage = (count*100)/df.shape[0]
percentage
print(f"there are {percentage.round(2)} percent chances that you will win the match after winning the toss")
```

# 2 Top Scorers

```python
df.head()
high = df.groupby("top_scorer")["highscore"].sum().sort_values(ascending = False).head(2)
high
```

```python
high.plot(kind = "barh")
```

# 10 Best Bowling Figures

```python
df["heighest_wickets"] = df["best_bowling_figure"].apply(lambda x: x.split("--")[0])
df["heighest_wickets"] = df["heighest_wickets"].astype(int)
top_bowlers = df.groupby("best_bowling")["heighest_wickets"].sum().sort_values(ascending = False).head(10)
top_bowlers.plot(kind = "barh")
```
